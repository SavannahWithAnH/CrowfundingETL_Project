### 2. Create the Campaign DataFrame
Extract and transform the crowdfunding.xlsx Excel data to create a campaign DataFrame has the following columns:


```python
#Data Extract
import pandas as pd
import regex as re
pd.options.display.max_colwidth = 200

```

```python
crowdfunding_df = pd.read_excel('../Resources/crowdfunding.xlsx')
crowdfunding_df
```


* The "cf_id" column - done
* The "contact_id" column - done
* The "company_name" column - done
* The "blurb" column, renamed to "description" - done
* The "goal" column, converted to the float data type - done
* The "pledged" column, converted to the float data type - done
* The "outcome" column - done
* The "backers_count" column - done
* The "country" column - done
* The "currency" column - done
* The "launched_at" column, renamed to "launch_date" and with the UTC times converted to the datetime format 
* The "deadline" column, renamed to "end_date" and with the UTC times converted to the datetime format
* The "category_id" column, with unique identification numbers matching those in the "category_id" column of the category DataFrame
* The "subcategory_id" column, with the unique identification numbers matching those in the "subcategory_id" column of the subcategory DataFrame

![image](https://github.com/SavannahWithAnH/Project-2/assets/126124356/a41de922-6bb9-41a2-aa76-3710624449e6)


```python
#Data Transfromation
campaign_df = crowdfunding_df[['cf_id','contact_id', 'company_name', 'blurb', 'goal', 'pledged', 'outcome', 'backers_count', 'country', 'currency', 'launched_at', 'deadline', 'category & sub-category']]
campaign_df = campaign_df.rename(columns= {'blurb': 'description', 'launched_at': 'launch_date', 'deadline': 'end_date'})
campaign_df['goal'] = campaign_df['goal'].astype(float)
campaign_df['pledged'] = campaign_df['pledged'].astype(float)
campaign_df["end_date"] = pd.to_datetime(campaign_df["end_date"], unit='s').dt.strftime('%Y-%m-%d')
campaign_df["launch_date"] = pd.to_datetime(campaign_df["launch_date"], unit='s').dt.strftime('%Y-%m-%d')
campaign_df[['category_id', 'subcategory_id']] = campaign_df['category & sub-category'].str.split(' ', n=1, expand=True)
campaign_df = campaign_df.drop('category & sub-category', axis=1)
campaign_df.head()
```


Export the campaign DataFrame as campaign.csv and save it to your GitHub repository.

```python
#Data Export
campaign_df.to_csv(('./Output/crowdfunding.csv'), index=False)
```
